```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

#create a simple series(avg, price in lakhs)
prices = pd.Series([10, 20, 30, 40, 50], name='Price in Lakhs')

#scikitlearn expects a 2D array, so we need to reshape/convert our 1D data series to 2D array
prices_reshaped = prices.values.reshape(-1, 1)

prices_reshaped
```

```python
#Initialize the scalers
std_scaler = StandardScaler()
minmax_scaler = MinMaxScaler()
robust_scaler = RobustScaler()
```

```python
# create a comparison dataframe
results = pd.DataFrame({
    'Original': prices,
    'Standarized (Z-score)' : std_scaler.fit_transform(prices_reshaped).flatten(),
    'Normalied (0-1)' : minmax_scaler.fit_transform(prices_reshaped).flatten(),
    'Robust (IQR)' : robust_scaler.fit_transform(prices_reshaped).flatten()   
})
print(results)
```
