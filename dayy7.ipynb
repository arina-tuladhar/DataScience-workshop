```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset("iris")
iris
```

```python
type(iris)
```

```python
iris['sepal_length']
```

```python
iris[['species', 'sepal_length']]
```

```python
corr = iris[['petal_length','petal_width']].corr()
corr
```

```python
print(iris[['sepal_length','sepal_width']].corr())
```

```python
iris.corr(numeric_only=True)
```

```python
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species')
plt.title('Petal Length vs Petal Width')
# plt.savefig('iris_scatterplot.png')  # Save the plot as a PNG file it must be saved before plt.show()
plt.show()
```

```python
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('sepal Length vs sepal Width')
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Iris Correlation Heatmap')
plt.show()
```
